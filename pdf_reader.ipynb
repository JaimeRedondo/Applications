## Import Libraries

```python
!pip3 install PyPDF2
! pip install langdetect

from langdetect import detect_langs 
from PyPDF2 import PdfReader
from textblob import TextBlob
import nltk
import os
import pandas as pd
import numpy as np
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')

cvs = os.listdir()[1:-1]
```

## Inputs

```python
lista_palabras = ["ENGLISH","FRENCH","SPANISH"]
```

## Function


```python
lista_cvs = cvs
def filtrar_cvs(lista_cvs,lista_palabras):

  porcentajes_matched = []
  lista_listas_matches = []
  lista_manual = []
  
  for cv in lista_cvs:
    try:
      reader = PdfReader(cv)
      text = ""
      for page in reader.pages:
        text += page.extract_text() + "\n" 
      print(cv)
     
      t = TextBlob(text)

      if str(detect_langs(text)[0])[0:2] == 'es':
        t = t.translate(from_lang='es', to='en')
      elif str(detect_langs(text)[0])[0:2] == 'it':
        t = t.translate(from_lang='it', to='en')
      elif str(detect_langs(text)[0])[0:2] == 'fr':
        t = t.translate(from_lang='fr', to='en')
      elif str(detect_langs(text)[0])[0:2] == 'en':
        t = t 
      else:
        print(f"El CV {cv} no está en un idioma muy utilizado")

      t = t.upper()

      lista_matches = []

      for palabra in lista_palabras:
        if palabra in t.words:
          lista_matches.append(palabra)
        else:
          pass

      matched = (len(lista_matches)/len(lista_palabras))*100
      porcentajes_matched.append(matched)
      lista_listas_matches.append(lista_matches)    

    except:
      print(f"el CV {cv} hay que mirarlo manualmente")
      lista_manual.append(cv) 
      porcentajes_matched.append('mirarlo manual')
      lista_listas_matches.append('mirarlo manual')

  #lista_cvs_updated = list(set(lista_cvs) - set(lista_manual))

  d = {'cv': lista_cvs, 'matched': porcentajes_matched, 'competencies_matched': lista_listas_matches}

  return pd.DataFrame(d)[pd.DataFrame(d)['matched']!='mirarlo manual'].sort_values('matched',ascending = False)
```

## Outputs

```python
filtrar_cvs(lista_cvs,lista_palabras)
```
